````python
import pandas as pd

def load_data(file_path):
    """
    Load data from a CSV file into a pandas DataFrame.

    Parameters:
    file_path (str): The path to the CSV file.

    Returns:
    pd.DataFrame: A DataFrame containing the loaded data.
    """
    try:
        data = pd.read_csv(file_path)
        return data
    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
        return None
    except pd.errors.EmptyDataError:
        print("Error: The file is empty.")
        return None
    except pd.errors.ParserError:
        print("Error: There was a parsing error while reading the file.")
        return None
````

````python
file_path  = '/Users/raman/Elnino/data/ERA5_single_levels_southasia/era5_single_levels_southasia_1980_2025_landcheck.csv'

# Solution: Use skipinitialspace=True to strip leading whitespace, 
# and specify na_values to treat whitespace as NaN
df = pd.read_csv(file_path, skipinitialspace=True, na_values=['', ' '])

# Strip column names too
df.columns = df.columns.str.strip()

print("=== Data types ===")
print(df.dtypes)

print("\n=== Null value counts for ALL columns ===")
null_counts = df.isna().sum()
print(null_counts)

print("\n=== Summary of columns ===")
print(f"Total columns: {len(df.columns)}")
print(f"Total rows: {len(df):,}")

# Check for columns that are incorrectly typed as strings (should be numeric)
print("\n=== Checking for string columns that should be numeric ===")
string_cols = df.select_dtypes(include=['object']).columns
for col in string_cols:
    print(f"\n{col}:")
    print(f"  - Type: {df[col].dtype}")
    print(f"  - Unique values: {df[col].nunique()}")
    print(f"  - Sample values: {df[col].dropna().unique()[:5].tolist()}")
    
    # Check if there are whitespace-only values
    if df[col].notna().any():
        whitespace_count = df[col].astype(str).str.strip().eq('').sum()
        print(f"  - Whitespace-only values: {whitespace_count}")

print("\n=== Checking numeric columns for unexpected patterns ===")
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    null_count = df[col].isna().sum()
    if null_count > 0:
        print(f"{col}: {null_count:,} nulls ({null_count/len(df)*100:.2f}%)")
    else:
        print(f"{col}: No nulls ✓")
````

## 🔍 Deep Dive: Hidden Missing Values Check
### Checking for disguised nulls, sentinel values, and data quality issues

````python
# Comprehensive check for hidden missing values in ERA5 South Asia file
import numpy as np

file_path_sa = '/Users/raman/Elnino/data/ERA5_single_levels_southasia/era5_single_levels_southasia_1980_2025_landcheck.csv'
df_sa = pd.read_csv(file_path_sa, skipinitialspace=True, na_values=['', ' '])
df_sa.columns = df_sa.columns.str.strip()

print("=" * 80)
print("COMPREHENSIVE HIDDEN MISSING VALUES CHECK")
print("=" * 80)

# 1. Check for common sentinel values (placeholder values for missing data)
print("\n1. CHECKING FOR SENTINEL VALUES (-999, -9999, 999, 9999, -1, etc.)")
print("-" * 80)

sentinel_values = [-999, -9999, -99, 999, 9999, 99, -1, -999.0, -9999.0]
numeric_cols = df_sa.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    sentinel_found = []
    for sentinel in sentinel_values:
        count = (df_sa[col] == sentinel).sum()
        if count > 0:
            sentinel_found.append(f"{sentinel}: {count} occurrences")
    
    if sentinel_found:
        print(f"\n⚠️ {col}:")
        for item in sentinel_found:
            print(f"   {item}")
    else:
        print(f"✅ {col}: No sentinel values found")

# 2. Check for infinity values
print("\n\n2. CHECKING FOR INFINITY VALUES")
print("-" * 80)

for col in numeric_cols:
    inf_count = np.isinf(df_sa[col]).sum()
    if inf_count > 0:
        print(f"⚠️ {col}: {inf_count} infinity values")
    else:
        print(f"✅ {col}: No infinity values")

# 3. Check for suspicious zeros (might be placeholders)
print("\n\n3. CHECKING FOR SUSPICIOUS ZEROS")
print("-" * 80)

for col in numeric_cols:
    zero_count = (df_sa[col] == 0).sum()
    zero_pct = (zero_count / len(df_sa)) * 100
    if zero_count > 0:
        print(f"{col}: {zero_count:,} zeros ({zero_pct:.2f}%)")
        # Show some context for zeros
        if col not in ['lsm']:  # lsm is expected to have zeros (ocean)
            sample_zeros = df_sa[df_sa[col] == 0][[col, 'latitude', 'longitude', 'lsm']].head(3)
            if len(sample_zeros) > 0:
                print(f"   Sample rows with zeros:")
                print(f"   {sample_zeros.to_string(index=False)}")
    else:
        print(f"✅ {col}: No zeros")

# 4. Check value ranges (physical plausibility)
print("\n\n4. CHECKING VALUE RANGES (Physical Plausibility)")
print("-" * 80)

# Expected ranges for climate variables
expected_ranges = {
    'sst': (200, 320),      # Sea surface temp in Kelvin (~-73°C to 47°C)
    't2m': (200, 330),      # 2m temperature in Kelvin
    'msl': (90000, 110000), # Mean sea level pressure in Pa
    'latitude': (-90, 90),
    'longitude': (-180, 360),
    'lsm': (0, 1),          # Land-sea mask fraction
}

for col, (min_exp, max_exp) in expected_ranges.items():
    if col in df_sa.columns:
        actual_min = df_sa[col].min()
        actual_max = df_sa[col].max()
        
        out_of_range = ((df_sa[col] < min_exp) | (df_sa[col] > max_exp)).sum()
        
        print(f"\n{col}:")
        print(f"   Expected range: [{min_exp}, {max_exp}]")
        print(f"   Actual range: [{actual_min:.2f}, {actual_max:.2f}]")
        
        if out_of_range > 0:
            print(f"   ⚠️ {out_of_range} values out of expected range!")
            outliers = df_sa[(df_sa[col] < min_exp) | (df_sa[col] > max_exp)][[col, 'latitude', 'longitude']].head(5)
            print(f"   Sample outliers:\n{outliers}")
        else:
            print(f"   ✅ All values within expected range")
````

````python
# Continue checking for hidden issues

# 5. Check for duplicate rows
print("\n\n5. CHECKING FOR DUPLICATE ROWS")
print("-" * 80)

duplicates = df_sa.duplicated().sum()
if duplicates > 0:
    print(f"⚠️ Found {duplicates:,} duplicate rows")
    # Show sample duplicates
    dup_samples = df_sa[df_sa.duplicated(keep=False)].head(6)
    print(f"Sample duplicates:\n{dup_samples}")
else:
    print(f"✅ No duplicate rows found")

# Check for duplicates in time-location combinations (should be unique)
time_loc_dups = df_sa.duplicated(subset=['time', 'latitude', 'longitude']).sum()
if time_loc_dups > 0:
    print(f"⚠️ Found {time_loc_dups:,} duplicate time-location combinations!")
else:
    print(f"✅ No duplicate time-location combinations")

# 6. Check for extreme outliers using IQR method
print("\n\n6. CHECKING FOR EXTREME OUTLIERS (using IQR method)")
print("-" * 80)

for col in numeric_cols:
    if col not in ['latitude', 'longitude', 'lsm']:  # Skip coords and binary
        # Calculate IQR
        Q1 = df_sa[col].quantile(0.25)
        Q3 = df_sa[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define outliers as values beyond 3*IQR
        lower_bound = Q1 - 3 * IQR
        upper_bound = Q3 + 3 * IQR
        
        outliers = ((df_sa[col] < lower_bound) | (df_sa[col] > upper_bound)).sum()
        outlier_pct = (outliers / len(df_sa)) * 100
        
        if outliers > 0:
            print(f"\n{col}:")
            print(f"   Q1={Q1:.2f}, Q3={Q3:.2f}, IQR={IQR:.2f}")
            print(f"   Bounds: [{lower_bound:.2f}, {upper_bound:.2f}]")
            print(f"   ⚠️ {outliers:,} outliers ({outlier_pct:.2f}%)")
        else:
            print(f"✅ {col}: No extreme outliers")

# 7. Check statistical summary
print("\n\n7. STATISTICAL SUMMARY")
print("-" * 80)
print(df_sa[numeric_cols].describe())
````

````python
# 8. Investigate the outliers found
print("\n\n8. INVESTIGATING OUTLIERS IN DETAIL")
print("-" * 80)

# Check SST outliers
print("\n📊 SST Outliers (below 296.43K or above 306.84K):")
Q1_sst = df_sa['sst'].quantile(0.25)
Q3_sst = df_sa['sst'].quantile(0.75)
IQR_sst = Q3_sst - Q1_sst
lower_sst = Q1_sst - 3 * IQR_sst
upper_sst = Q3_sst + 3 * IQR_sst

sst_outliers = df_sa[(df_sa['sst'] < lower_sst) | (df_sa['sst'] > upper_sst)][['time', 'latitude', 'longitude', 'sst', 'lsm', 'land_ocean']]
if len(sst_outliers) > 0:
    print(f"   Total: {len(sst_outliers)} outliers")
    print(f"   Min SST outlier: {sst_outliers['sst'].min():.2f}K ({sst_outliers['sst'].min()-273.15:.2f}°C)")
    print(f"   Max SST outlier: {sst_outliers['sst'].max():.2f}K ({sst_outliers['sst'].max()-273.15:.2f}°C)")
    print(f"\n   Sample outliers:")
    print(sst_outliers.head(10))
    
    # Are these outliers associated with specific locations?
    print(f"\n   Distribution by land_ocean category:")
    print(sst_outliers['land_ocean'].value_counts())

# Check wind component outliers
print("\n\n📊 Wind Component Outliers (avg_iews):")
Q1_iews = df_sa['avg_iews'].quantile(0.25)
Q3_iews = df_sa['avg_iews'].quantile(0.75)
IQR_iews = Q3_iews - Q1_iews
lower_iews = Q1_iews - 3 * IQR_iews
upper_iews = Q3_iews + 3 * IQR_iews

iews_outliers = df_sa[(df_sa['avg_iews'] < lower_iews) | (df_sa['avg_iews'] > upper_iews)][['time', 'latitude', 'longitude', 'avg_iews', 'avg_inss']]
if len(iews_outliers) > 0:
    print(f"   Total: {len(iews_outliers)} outliers")
    print(f"   Min: {iews_outliers['avg_iews'].min():.4f}")
    print(f"   Max: {iews_outliers['avg_iews'].max():.4f}")
    print(f"\n   Sample outliers:")
    print(iews_outliers.head(10))

# 9. Check for constant values (columns that never change)
print("\n\n9. CHECKING FOR CONSTANT/NEAR-CONSTANT COLUMNS")
print("-" * 80)

for col in df_sa.columns:
    if col not in ['time']:  # Skip time
        unique_count = df_sa[col].nunique()
        unique_pct = (unique_count / len(df_sa)) * 100
        
        if unique_count == 1:
            print(f"⚠️ {col}: CONSTANT - only 1 unique value: {df_sa[col].unique()[0]}")
        elif unique_count < 10:
            print(f"⚠️ {col}: NEAR-CONSTANT - only {unique_count} unique values")
        elif unique_pct < 1:
            print(f"   {col}: Low diversity - {unique_count} unique values ({unique_pct:.2f}%)")
        else:
            print(f"✅ {col}: Good diversity - {unique_count} unique values ({unique_pct:.2f}%)")
````
